```python
import matplotlib.pyplot as plt
import numpy as np

models = ["Random Grouping", "BERT-Zero-Shot", "DeBERTa-Zero-Shot", "DeBERTa-Few-Shot (k=20)"]
mean_swaps = [4.47, 4.18 , 3.76, 4.00]  

x = np.arange(len(models))
width = 0.6

fig, ax = plt.subplots(figsize=(10,5))
bars = ax.bar(x, mean_swaps, width, color=["blue", "green", "red", "purple"])

for bar in bars:
    height = bar.get_height()
    ax.annotate(f'{height:.2f}',
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 3), 
                textcoords="offset points",
                ha='center', va='bottom')

ax.set_ylabel("Mean Min Swaps")
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.set_title("Comparison of Models by Mean Min Swaps")

plt.show()
```
