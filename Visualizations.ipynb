```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

models = ["Random Grouping", "BERT-Zero-Shot", "DeBERTa-Zero-Shot", "DeBERTa-Few-Shot (k=20)"]
mean_swaps = [4.47, 4.18 , 3.76, 4.00]  

x = np.arange(len(models))
width = 0.6

fig, ax = plt.subplots(figsize=(10,5))
bars = ax.bar(x, mean_swaps, width, color=["blue", "green", "red", "purple"])

for bar in bars:
    height = bar.get_height()
    ax.annotate(f'{height:.2f}',
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 3), 
                textcoords="offset points",
                ha='center', va='bottom')

ax.set_ylabel("Mean Min Swaps")
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.set_title("Comparison of Models by Mean Min Swaps")

plt.show()
```

```python
df_results = pd.read_csv("reports/LLM_few_shot_results.csv")

plt.figure()
plt.plot(df_results["k"], df_results["zero_one_accuracy"], marker="o")

for x, y in zip(df_results["k"], df_results["zero_one_accuracy"]):
    plt.text(x, y + 0.02, f"k = {x}", ha="center")
plt.xlabel("k")
plt.ylabel("Zero-One Accuracy")
plt.title("LLM: Zero-One Accuracy vs k")
plt.show()

plt.figure()
plt.plot(df_results["k"], df_results["mean_swaps"], marker="o")

for x, y in zip(df_results["k"], df_results["mean_swaps"]):
    plt.text(x, y + 0.02, f"k = {x}", ha="center")
plt.xlabel("k")
plt.ylabel("Mean Swaps")
plt.title("LLM: Mean Swaps vs k")
plt.show()

plt.figure()
plt.plot(df_results["k"], df_results["runtime"], marker="o")

for x, y in zip(df_results["k"], df_results["runtime"]):
    plt.text(x, y + 0.02, f"k = {x}", ha="center")

plt.ylim(bottom=0)
plt.xlabel("k")
plt.ylabel("Runtime (seconds)")
plt.title("LLM: Runtime vs k")
plt.show()
```
